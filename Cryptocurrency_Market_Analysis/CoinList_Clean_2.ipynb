```python
import pandas as pd
```

```python
cp1 = pd.read_csv('/content/CoinPrices_CleanFinal.csv')
```

```python
cp1.head(4)
```

```python
cp1['Date']= pd.to_datetime(cp1['Date'])
```

```python
cp1['DateDiff']=cp1.groupby('CoinName')['Date'].transform(lambda x: (x.max() - x.min()).days)
```

```python
cp1.info()
```

```python
cp1.head(5)
```

```python
cp1.sample(5)
```

```python
cp1_onlyDateDiff = cp1.iloc[:,7:9]
```

```python
cp1_onlyDateDiff = cp1_onlyDateDiff.drop_duplicates(subset=['CoinName'])
```

```python
cp1_onlyDateDiff.head(3)
```

```python
cl1 = pd.read_csv('/content/CoinList_CleanFinal.csv')
```

```python
cl1
```

```python
cp1_onlyDateDiff = cp1_onlyDateDiff.rename(columns={'CoinName':'Name'})
```

```python
cp1_onlyDateDiff.head(3)
```

```python
cl1_add_DateDiff = pd.merge(cp1_onlyDateDiff,cl1, how = 'outer', on = 'Name')
```

```python
cl1_add_DateDiff['DateDiff'] = cl1_add_DateDiff['DateDiff'].fillna(0)
```

```python
cl1_add_DateDiff['DateDiff'] = cl1_add_DateDiff['DateDiff'].astype(int)
```

```python
cl1_add_DateDiff
```

```python
cl1_add_DateDiff.info()
```

```python
cl1_add_DateDiff.to_csv('New_CoinList_Clean.csv',index = False)
```

```python
cl2= pd.read_csv('/content/New_CoinList_Clean.csv')
```

```python
def Check(x):
    if x['DateDiff'] < 90 and x['Volume24h'] == 0:
        val = 'Scam'
    else:
        val = 'Legit'
    return val
```

```python
cl2['Check'] = cl2.apply(Check, axis=1)
```

```python
cl2.groupby('Check').agg(Coin=('Name','nunique')).reset_index()
```
