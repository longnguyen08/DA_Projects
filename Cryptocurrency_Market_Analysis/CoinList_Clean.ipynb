```python
import pandas as pd
```

```python
cl = pd.read_csv('/content/CoinList.csv')
```

```python
cl.head(5)
```

```python
cl1 = cl.iloc[:,3:11]
```

```python
cl1.head(4)
```

```python
cl1 = cl1.rename(columns={'Symbol' : 'Name', 'Market.Cap': 'MarketCap', 'Circulating.Supply': 'Circulating_Supply', 'Volume..24h.' : 'Volume24h', 'X..1h':'X_1h','X..24h':'X_24h','X..7d':'X_7d'})
```

```python
cl1.head(4)
```

```python
cl1.isnull().sum()
```

```python
cl1.info()
```

```python
cl1 = cl1.drop_duplicates(subset=['Name'])
```

```python
cl1['MarketCap'] = cl1['MarketCap'].transform(lambda x: x.replace('$',''))
cl1['MarketCap'] = cl1['MarketCap'].transform(lambda x: x.replace(',',''))
cl1['MarketCap'] = cl1['MarketCap'].transform(lambda x: x.replace('?','0'))
cl1['MarketCap'] = cl1['MarketCap'].astype(int)
```

```python
cl1['Price'] = cl1['Price'].transform(lambda x: x.replace('?','0'))
cl1['Price'] = cl1['Price'].transform(lambda x: x.replace('$',''))
```

```python
cl1.sample(5)
```

```python
cl1['Circulating_Supply'] = cl1['Circulating_Supply'].transform(lambda x: x.replace('?','0'))
cl1['Circulating_Supply'] = cl1['Circulating_Supply'].transform(lambda x: x.replace('\n*',''))
cl1['Circulating_Supply'] = cl1['Circulating_Supply'].transform(lambda x: x.replace('\n *',''))
cl1['Circulating_Supply'] = cl1['Circulating_Supply'].transform(lambda x: x.replace(',',''))
```

```python
cl1['Volume24h'] = cl1['Volume24h'].transform(lambda x: x.replace('$',''))
cl1['Volume24h'] = cl1['Volume24h'].transform(lambda x: x.replace(',',''))
cl1['Volume24h'] = cl1['Volume24h'].transform(lambda x: x.replace('Low Vol','0'))
cl1['Volume24h'] = cl1['Volume24h'].transform(lambda x: x.replace('?','0'))
cl1['Volume24h'] = cl1['Volume24h'].astype(int)
```

```python
cl1.sample(5)
```

```python
cl1['X_1h'] = cl1['X_1h'].transform(lambda x: x.replace('Low Vol','0.00'))
cl1['X_1h'] = cl1['X_1h'].transform(lambda x: x.replace('?','0.00'))
cl1['X_1h'] = cl1['X_1h'].transform(lambda x: x.replace('%',''))
cl1['X_1h'] = cl1['X_1h'].astype(float)
```

```python
cl1['X_24h'] = cl1['X_24h'].transform(lambda x: x.replace('Low Vol','0.00'))
cl1['X_24h'] = cl1['X_24h'].transform(lambda x: x.replace('?','0.00'))
cl1['X_24h'] = cl1['X_24h'].transform(lambda x: x.replace('%',''))
cl1['X_24h'] = cl1['X_24h'].astype(float)
```

```python
cl1['X_7d'] = cl1['X_7d'].transform(lambda x: x.replace('Low Vol','0.00'))
cl1['X_7d'] = cl1['X_7d'].transform(lambda x: x.replace('?','0.00'))
cl1['X_7d'] = cl1['X_7d'].transform(lambda x: x.replace('%',''))
cl1['X_7d'] = cl1['X_7d'].astype(float)
```

```python
cl1.info()
```

```python
cl1.to_csv('CoinList_Clean.csv',index = False)
```
